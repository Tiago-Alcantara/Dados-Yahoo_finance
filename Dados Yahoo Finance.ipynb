## Leitura e extração de dados

```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
#setando a ação escolhida e puxando ela
acao = "WEGE3.SA"

dados = yf.download(tickers= acao)

dados
```

```python
#puxando ação determinando alguns parametros como desde tal data ate tal data

acao = "WEGE3.SA"

dados = yf.download(tickers= acao, start="2010-12-31", end = datetime.now())

dados
```

```python
acao = "WEGE3.SA"

dados = yf.download(tickers= acao, start="2010-12-31",end=datetime.now(),
                        interval= "1wk")# intervalo semanal 

#pode inventar moda, porem pra qualquer dado espere para se fazer no pandas 

dados
```

```python
#puxar dados várias ações
#criar lista 
acao = ["META", "NKE", "GOOGL", "AMZN"]

dados = yf.download(tickers= acao, start = "2010-12-31", end = datetime.now())

dados
```

```python
#dados['Adj Close']['META']

dados['Adj Close']
```

```python
#puxar dados várias ações
acao = ["RPAD5.SA", "CASH3.SA", "VALE3.SA", "ATOM3.SA"]

dados = yf.download(tickers= acao,
                        start = "2018-12-31",end = datetime.now())['Adj Close']

dados.head(50)
```
